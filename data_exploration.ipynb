```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd

books = pd.read_csv(f"{path}/books.csv")
```

```python
# iterating through each column to check the unique values

for column in books.columns:
    unique_values = books[column].unique()
    print(f"Column '{column}': {unique_values}")
    print(f"Number of unique values: {len(unique_values)}")
    print("-" * 50)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(),cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing")

plt.show()
```

```python
books["missing_description"] = np.where(books["description"].isna(),1,0)
books["age_of_book"] = 2024 - books["published_year"]

columns_of_interest = ["num_pages","age_of_book","missing_description","average_rating"]

correlation_matrix = books[columns_of_interest].corr(method = "spearman")

sns.set_theme(style="white")
plt.figure(figsize=(10,10))
heatmap = sns.heatmap(correlation_matrix, annot = True, fmt = ".2f", cmap = "coolwarm",
                     cbar_kws={"label": "Spearman Correlation"})
heatmap.set_title("Correlation Heatmap")
plt.show()
```

There seems to be no correlation between missing description entries and anything else.

And since there are only a few missing descriptions, it's easier to just delete those entries

```python
# calculating the number of missing rows for each of the important columns
books[(books["description"].isna()) |
      (books["num_pages"].isna()) |
      (books["average_rating"].isna()) |
      (books["published_year"].isna())
]
```

```python
# creating a new dataframe without all those 303 missing value rows
books_final = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["published_year"].isna())
]
```

```python
books_final
```

## Analysing the description column

There are some rows that have descriptions that either have way less information or irrelevant information. Based on the number of these rows, they should be dealt with/deleted.

```python
# creating a new column to calculate thew number of words in each row's description column
books_final["num_of_words_in_description"] = books_final["description"].str.split().str.len()
```

```python
# analyzing the descriptions that have between 1 and 4 words.
books_final.loc[books_final["num_of_words_in_description"].between(1,4),"description"] 
```

```python
# analyzing the descriptions that have between 5 and 14 words.
books_final.loc[books_final["num_of_words_in_description"].between(5,14),"description"] 
```

```python
# analyzing the descriptions that have between 15 and 24 words.
books_final.loc[books_final["num_of_words_in_description"].between(15,24),"description"] 
```

```python
# analyzing the descriptions that have between 25 and 34 words.
books_final.loc[books_final["num_of_words_in_description"].between(25,34),"description"] 
```

It seems like, the descriptions make more sense and are more relevant when they have 25 words or more. Therefore, the ones with less than 25 words will be deleted.

```python
books_final_25_words = books_final[books_final["num_of_words_in_description"] >= 25]
```

```python
books_final_25_words
```

```python
books_final_25_words["tagged_description"] = books_final_25_words[["isbn13","description"]].astype(str).agg(" ".join, axis = 1)
```

The dataframe has been cleaned fully. Now we will delete some useless columns and then save the cleaned data to be used later.

```python
(
    books_final_25_words
    .drop(["missing_description","age_of_book","num_of_words_in_description"], axis = 1)
    .to_csv("books_cleaned.csv", index = False)
)
```
